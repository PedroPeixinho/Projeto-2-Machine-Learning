### 2.1 Descri√ß√£o das M√©tricas Utilizadas

No contexto da avalia√ß√£o de modelos de Machine Learning, v√°rias m√©tricas s√£o utilizadas para medir a efic√°cia e a precis√£o dos modelos. Entre as mais comuns e relevantes est√£o a acur√°cia, o F1-Score, a curva ROC, recall e precis√£o.

1. **Acur√°cia**:
   - **Defini√ß√£o**: A acur√°cia √© a propor√ß√£o de previs√µes corretas (tanto verdadeiros positivos quanto verdadeiros negativos) em rela√ß√£o ao total de previs√µes feitas pelo modelo.
   - **Import√¢ncia**: A acur√°cia √© uma m√©trica simples e intuitiva, √∫til quando as classes s√£o balanceadas. No entanto, em casos de classes desbalanceadas, pode ser enganosa, pois um modelo que prev√™ sempre a classe majorit√°ria pode ter alta acur√°cia, mas baixa efic√°cia na detec√ß√£o da classe minorit√°ria.

2. **F1-Score**:
   - **Defini√ß√£o**: O F1-Score √© a m√©dia harm√¥nica entre a precis√£o (precision) e o recall, equilibrando a rela√ß√£o entre falsos positivos e falsos negativos.
   - **Import√¢ncia**: O F1-Score √© especialmente √∫til em situa√ß√µes de classes desbalanceadas, pois considera tanto a precis√£o quanto o recall, proporcionando uma vis√£o mais equilibrada do desempenho do modelo.

3. **Curva ROC e AUC (Area Under the Curve)**:
   - **Defini√ß√£o**: A curva ROC (Receiver Operating Characteristic) √© um gr√°fico que representa a taxa de verdadeiros positivos (TPR) contra a taxa de falsos positivos (FPR) para diferentes limiares de decis√£o. A AUC (√Årea Sob a Curva) quantifica a √°rea total sob a curva ROC.
   - **Import√¢ncia**: A curva ROC e a AUC s√£o √∫teis para avaliar a capacidade de discrimina√ß√£o do modelo entre classes. Uma AUC de 1 indica um modelo perfeito, enquanto uma AUC de 0,5 indica um modelo que n√£o tem capacidade discriminativa, equivalente a uma classifica√ß√£o aleat√≥ria.

4. **Recall**:
   - **Defini√ß√£o**: O recall, tamb√©m conhecido como sensibilidade ou taxa de verdadeiros positivos, √© a propor√ß√£o de verdadeiros positivos em rela√ß√£o ao total de casos positivos reais.
  
   - **Import√¢ncia**: O recall √© crucial em aplica√ß√µes onde a detec√ß√£o dos casos positivos √© fundamental, como em diagn√≥sticos m√©dicos ou em sistemas de seguran√ßa, onde a falha em identificar positivos pode ter consequ√™ncias graves.

5. **Precis√£o**:

  - **Defini√ß√£o**: A precis√£o (precision) √© a propor√ß√£o de verdadeiros positivos em rela√ß√£o ao total de previs√µes positivas feitas pelo modelo.

  - **Import√¢ncia**: A precis√£o √© importante em contextos onde o custo dos falsos positivos √© alto. Essas m√©tricas s√£o essenciais para uma avalia√ß√£o abrangente do desempenho dos modelos de Machine Learning, permitindo identificar e corrigir poss√≠veis defici√™ncias, al√©m de garantir que o modelo selecionado atenda adequadamente aos objetivos do projeto.

#### 2.3 Implementa√ß√£o e resultados do teste estat√≠stico de compara√ß√£o de classificadores de Janez Demsar.

Foi escolhido o teste de Friedmann para realizar a compara√ß√£o dos classificadores.

Para o teste de Friedman ser√£o testadas as seguintes hip√≥teses, de acordo com o f1 score de cada modelo:

H0 (hip√≥tese nula): Os classificadores t√™m resultados similares.

H1 (hip√≥tese alternativa): Os classificadores t√™m resultados diferentes.

O teste de Friedman testar√° se a hip√≥tese nula ser√° rejeitada ou n√£o de acordo com o valor cr√≠tico de Œ± = 0.05. Caso ela seja negada, ser√° realizado o teste de Nemenyi para verificar quais classificadores t√™m resultados diferentes.

Para o teste, foi utilizado o cross validation em 5 folds, onde foram calculados os f1 score de cada fold para cada modelo.

```python
# Classificadores
classifiers = [best_knn, best_lvq, best_dt, best_svm, best_rf, best_mlp, committee_best, heterogeneous_committee_best]

# N√∫mero de folds
num_folds = 5

# Inicializa o objeto KFold
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Lista dos f1_scores
f1_scores = np.zeros((num_folds, len(classifiers)))

# Separa o dataset de valida√ß√£o em folds e avalia o f1_score para cada fold, para cada modelo
for fold, (train_index, test_index) in enumerate(kf.split(cleaned_val_df_scaled)):
    X_test_fold, y_test_fold = cleaned_val_df_scaled.iloc[test_index], val_target.iloc[test_index]

    for idx, clf in enumerate(classifiers):
        y_pred = clf.predict(X_test_fold)
        f1 = f1_score(y_test_fold, y_pred, average='weighted')
        f1_scores[fold, idx] = f1
```

```python
from scipy.stats import friedmanchisquare
from scikit_posthocs import posthoc_nemenyi_friedman

stat, p = friedmanchisquare(*f1_scores.T)
print(f"Friedman test statistic: {stat}, p-value: {p}")

if p < 0.05:
  nemenyi_results = posthoc_nemenyi_friedman(f1_scores)
  print("Nemenyi test results:\n", nemenyi_results)
else:
  print("No significant differences found with Friedman test.")
```

Como o valor de p-value foi menor que o valor cr√≠tico, a hip√≥tese nula foi rejeitada e assim os classificadores t√™m resultados diferentes.

Pelo teste de Nemenyi, pode-se perceber que existem classificadores com p-value menor que o valor cr√≠tico entre si, sendo eles os que possuem os resultados divergentes para o f1 score. Nesse caso, a diverg√™ncia ocorreu entre:
